```python
#Torres de Hanoi - Divide y venceras
def resolver(N, desde, hasta):
  #N - N de fichas
  #desde - torre inicial
  #hasta - torre fina
  if N==1 :
    print("Lleva la ficha desde " + str(desde) + " hasta " + str(hasta))

  else:
    resolver(N-1, desde, 6-desde-hasta)
    print("Lleva la ficha desde " + str(desde) + " hasta " + str(hasta))
    resolver(N-1, 6-desde-hasta,  hasta)

resolver(4, 1, 3)
# Observaciones:
# columna 1, 2 y 3 desde, hasta y pivote (cualquiera de ellos)
# 1 + 2 + 3 = 6 (cul es el pivote? 6 - desde - hasta)
# Operaciones:
# T(1)=1
# T(n) = 1 + 2 T(N-1)
#      = 1 + 2(1 + 2T(n-2))
#
# Orden (2 elevado a n)
# O(2^n)
```

```python
#N Reinas - Vuelta Atrs()
################################################################################

#Verifica que en la solucin parcial no hay amenzas entre reinas
################################################################
def es_prometedora(SOLUCION,etapa):
################################################################
  #print(SOLUCION)
  #Si la solucin tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " est " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True

#Traduce la solucin al tablero
################################################################
def escribe_solucion(S):
################################################################
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")



#Proceso principal de N-Reinas
################################################################
def reinas(N, solucion=[],etapa=0):
################################################################
### ....
  if len(solucion) == 0:         # [0,0,0...]
    solucion = [0 for i in range(N) ]
  #print(solucion)
  for i in range(1, N+1):
    solucion[etapa] = i
    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

  solucion[etapa] = 0

reinas(8,solucion=[],etapa=0)

# O(n^2)
# Analizar y evaluar
```

```python
# Problema: Encontrar los dos puntos ms cercanos
# Dado un conjunto de n puntos se trata de encontrar los dos puntos mas cercanos (menor distancia entre si)
# Fuerza bruta: Explorando todas la posibles soluciones
%%time
def calcular_distancia(a,b):
  return abs(a-b)

distancia_minima = float("inf")
for i in range(len(lista_1d)):
  for j in range(i+1, len(lista_1d)):
    distancia = calcular_distancia(lista_1d[i], lista_1d[j])
    if distancia < distancia_minima:
      distancia_minima = distancia
      puntos_cercanos = (lista_1d[i], lista_1d[j])

print(distancia_minima, puntos_cercanos)
```

```python
# Observaciones:
# El orden del algoritmo es O(n^2) porque necesitamos comparar la distancia de cada punto con los otros.
# El primer punto compararlo con n-1 puntos, el segundo con n-2 y as sucesivamente,
# si sumamos todas las comparaciones (n-1) + (n+2) + ...
# Orden = (n*(n-1))/2 ==> O(n^2)
# Se puede Mejorar? Si, puede mejorarse el c[odigo para hacerse ms entendible.
# Pero el algoritmo seguir teniendo el mismo comportamiento en trminos de orden de ejecucon.
```

```python
%%time
# Divide y vencers: dividir el conjunto en dos mitades, resolver recursivamente en cada mitad y
# combinar los resultados revisando si el par mas cercano cruza la frontera entre mitades.
def closest_pair_1d(points):
  points = sorted(points)
  return divide_and_conquer(points)

def divide_and_conquer(arr):
  n = len(arr)

  # Caso base Calcular todos los pares
  if n <= 3:
    min_dist = float("inf")
    best_pair = None
    for i in range(n):
      for j in range(i+1, n):
        d = abs(arr[i] - arr[j])
        if d < min_dist:
          min_dist = d
          best_pair = (arr[i], arr[j])
    return min_dist, best_pair
  # Dividir
  mid = n//2
  left = arr[:mid]
  right = arr[mid:]
  # Resolver recursivamente
  d_left, pair_left = divide_and_conquer(left)
  d_right, pair_right = divide_and_conquer(right)
  # Mejor distancia parcial
  if d_left < d_right:
    d = d_left
    best_pair = pair_left
  else:
    d = d_right
    best_pair = pair_right

  # Revisar cruce entre mitades
  d_cross = abs(left[-1] - right[0])
  if d_cross < d:
    return d_cross, (left[-1], right[0])

  return d, best_pair

print(lista_1d)
distancia, puntos = closest_pair_1d(lista_1d)
print("distancia minima:", distancia, "puntos:", puntos)
```

```python
# El orden del algoritmo es O(nlogn)
# Ordenamiento O(nlogn)
# Divide y venceras O(n)
# O(nlogn)
```

```python
 # Extender el algoritmo a 2D:[(1122,6175),(135,4076),(70296,2731)..]
 # El algoritmo sigue la misma estructura:
 # 1. Ordenar los puntos por coordenada x.
 # 2. Dividir en dos mitades.
 # 3. Conquistar recursivamente cada mitad.
 # 4. Combinar revisando pares que cruzan la frontera. Aqu est la clave, en 2D no basta con revisar un solo par cruzado como en 1D.
```

```python

# En 1D, si un par cruza la frontera, necesariamente es: "ltimo de la izquierda    primero de la derecha"
# En 2D, un punto de la izquierda puede estar muy cerca de varios puntos de la derecha, no solo del primero.
%%time
import random
import math # librera matemtica

def distancia(p1, p2):
    return math.dist(p1, p2)  # Euclidean distance between two points p and q given as an iterable of coordinates

def closest_pair_2d(points):
    points = sorted(points, key=lambda p: p[0]) # Ordenado por coordenada x, para dividir el pano en 2
    return divide_and_conquer(points)

def divide_and_conquer(P):
    n = len(P)

    # Caso base
    if n <= 3:
        min_d = float("inf")
        best_pair = None
        for i in range(n):
            for j in range(i+1, n):
                d = distancia(P[i], P[j])
                if d < min_d:
                    min_d = d
                    best_pair = (P[i], P[j])
        return min_d, best_pair

    mid = n // 2
    mid_x = P[mid][0]

    L = P[:mid]
    R = P[mid:]

    dL, parL = divide_and_conquer(L)
    dR, parR = divide_and_conquer(R)

    if dL < dR:
        d = dL
        best_pair = parL
    else:
        d = dR
        best_pair = parR

    # Construir la franja, vertical alrededor de la linea divisoria
    # d unidades hacia la izquierda + d unidades hacia la derecha = 2d unidades de ancho para la franja
    # Para que un par cruzado pueda tener distancia < d, la diferencia en x entre los dos puntos debe ser < d.
    strip = [p for p in P if abs(p[0] - mid_x) < d]
    strip.sort(key=lambda p: p[1])

    # Comparar puntos en la franja
    for i in range(len(strip)):
        for j in range(i+1, min(i+7, len(strip))):
            d_ij = distancia(strip[i], strip[j])
            if d_ij < d:
                d = d_ij
                best_pair = (strip[i], strip[j])

    return d, best_pair

lista_2d = [(random.randrange(1,1000000),random.randrange(1,1000000)) for x in range(10000)]
print(lista_2d)
dist, par = closest_pair_2d(lista_2d)
print(dist, par)

```

```python
# El orden del algoritmo es O(nlogn)
# Ordenamiento O(nlogn)
# Divide y venceras O(n)
# O(nlogn)
```

```python
# Extender el algoritmo a 3D
# [(x1, y1, z1), (x2, y2, z2), ...]
# d(p,q) = SQRT((xp-xq)^2+(yp-yq)^2+(zp-zq)^2)
# 1. Ordenar por coordenada x
# 2. Dividir en dos mitades
# 3. Resolver recursivamente
# 4. Combinar revisando pares que cruzan la frontera, la diferencia est en cmo se construye y procesa la franja.
import random
import math

def dist3(p1, p2):
    return math.dist(p1, p2)

def closest_pair_3d(points):
    points = sorted(points, key=lambda p: p[0])
    return divide_and_conquer(points)

def divide_and_conquer(P):
    n = len(P)

    # Caso base
    if n <= 3:
        min_d = float("inf")
        best_pair = None
        for i in range(n):
            for j in range(i+1, n):
                d = dist3(P[i], P[j])
                if d < min_d:
                    min_d = d
                    best_pair = (P[i], P[j])
        return min_d, best_pair

    mid = n // 2
    mid_x = P[mid][0]

    L = P[:mid]
    R = P[mid:]

    dL, parL = divide_and_conquer(L)
    dR, parR = divide_and_conquer(R)

    if dL < dR:
        d = dL
        best_pair = parL
    else:
        d = dR
        best_pair = parR

    # Construir franja 3D
    strip = [p for p in P if abs(p[0] - mid_x) < d]
    strip.sort(key=lambda p: p[1])  # ordenar por y

    # Comparar puntos cercanos en la franja.
    # En 3D, cada punto de la franja necesita compararse con los siguientes 1520 puntos.
    for i in range(len(strip)):
        for j in range(i+1, min(i+20, len(strip))):  # ~20 vecinos
            d_ij = dist3(strip[i], strip[j])
            if d_ij < d:
                d = d_ij
                best_pair = (strip[i], strip[j])

    return d, best_pair

lista_3d = [(random.randrange(1,1000000),random.randrange(1,1000000),random.randrange(1,1000000)) for x in range(10000)]
print(lista_3d)
distancia, par = closest_pair_3d(lista_3d)
print("Distancia mnima:", distancia)
print("Puntos ms cercanos:", par)
```

```python
# El orden del algoritmo es O(nlogn)
# Ordenamiento O(nlogn)
# Divide y venceras O(n)
# O(nlogn)
```

---
El algoritmo de Divide y Vencers para el problema del par de puntos ms cercanos fue introducido formalmente por **Bentley y Shamos (1976)** y refinado en trabajos posteriores como **Shamos y Hoey (1975)** y la revisin moderna de **Smid (2002)**.  
Una de sus ideas centrales es que, tras dividir el conjunto en dos mitades y obtener la distancia mnima parcial \( \delta \), solo un nmero **constante** de puntos en la franja deben compararse.

La franja se define como la regin:

$$ \{ p \in \mathbb{R}^d : |p_x - x_m| < \delta \} $$

- \( p \) es un punto en dimensin \( d \)
- \( p_x \) es su coordenada respecto al eje de divisin
- \( x_m \) es la coordenada del plano (o punto) medio
- \( \delta \) es la distancia mnima parcial obtenida en las mitades

---

# 1D: Intervalo en la recta

### Definicin geomtrica
En una dimensin, los puntos estn sobre una recta. La franja es simplemente el intervalo:

$$
[x_m - \delta,\; x_m + \delta]
$$


### Propiedad clave
En 1D, si los puntos estn ordenados, **dos puntos ms cercanos deben ser vecinos en el ordenamiento**.  
Esto implica que dentro de la franja solo es necesario comparar **cada punto con su siguiente**.

### Justificacin terica
Bentley y Shamos muestran que en 1D el problema se reduce a comparar pares adyacentes, ya que cualquier punto ms lejano que el vecino inmediato tiene distancia mayor.

### Complejidad
$$
O(n)
$$

---

# 2D: Banda vertical y el lmite de 6 puntos

### Definicin geomtrica
En 2D, la franja es una banda vertical de ancho \( 2\delta \).  
Los puntos dentro de la franja se ordenan por su coordenada \( y \).

### Teorema fundamental (BentleyShamos)
Cada punto de la franja solo necesita compararse con los **siguientes 6 puntos** en el orden por \( y \).

### Justificacin geomtrica
La demostracin se basa en un argumento de **empaquetamiento de discos**:

- Si un punto tiene distancia mnima \( \delta \), entonces en un cuadrado de lado \( \delta \) no pueden caber ms de 6 puntos sin violar la distancia mnima.
- Por lo tanto, al ordenar por \( y \), solo los puntos cercanos en esa dimensin pueden estar a distancia menor que \( \delta \).

Este resultado aparece formalmente en:

- Bentley & Shamos (1976)  
- Smid (2002), seccin *Closest Point Problems*

### Complejidad
$$
O(n)
$$

---

# 3D: Prisma y vecinos en un cubo de lado \( \delta \)

### Definicin geomtrica
En 3D, la franja es un **prisma rectangular** de ancho \( 2\delta \).  
Los puntos se ordenan por \( y \) y \( z \), o se agrupan en celdas cbicas.

### Propiedad clave
En 3D, cada punto solo necesita compararse con los puntos que caen dentro de un **cubo de lado \( \delta \)** alrededor de l.

### Justificacin geomtrica
La demostracin usa empaquetamiento de **esferas**:

- En un cubo de lado \( \delta \), solo puede haber un nmero **constante** de puntos separados al menos \( \delta \).  
- Este nmero es tpicamente  26 (los vecinos de un voxel 3D), aunque la constante exacta no afecta la complejidad.

Este resultado est formalizado en:

- Shamos & Hoey (1975)  
- Smid (2002), captulo del *Handbook of Computational Geometry*

### Complejidad
$$
O(n)
$$
